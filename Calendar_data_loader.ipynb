```python
from bs4 import BeautifulSoup

for i in range(len(li)-1):
  print(i)
  soup = BeautifulSoup(li[i], 'html.parser')

  rect_attributes = []
  for rect_tag in soup.find_all('rect'):
      attributes = {
          'professor': professors[i],
          'x': rect_tag.get('x'),
          'y': rect_tag.get('y'),
          'height': rect_tag.get('height'),
          'text':rect_tag.text
      }
      rect_attributes.append(attributes)

  # Print the result
  for attributes in rect_attributes:
      print(attributes)
```

```python
import pandas as pd
# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(rect_attributes)

# Print the resulting DataFrame
df
```
